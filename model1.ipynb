```python
#Filtered data for Data Preprocessing
FD = data[['C1','C2','C3Prime','C4','C5Prime','C6Prime','C7','T1Prime','T2Prime','S1Prime','S2Prime','S3Prime','S4Prime','Y1']].copy()
#Displaying the new dataframe FD
FD.dtypes
```

```python
#Checking for missing values since now the dataset is ready 
#FD.isnull().sum(axis=0).tolist()
#Create a for loop to count the missing values in each column in the dataset 
names= []
val = []
for col in FD.columns:
    names.append(col)
    val.append(FD[col].isnull().sum())
    print(col, val[-1])
```

```python
FD['T1Prime'] = pd.to_numeric(FD.T1Prime,errors='coerce' )
FD['T2Prime'] = pd.to_numeric(FD.T2Prime,errors='coerce' )
FD['S1Prime'] = pd.to_numeric(FD.S1Prime,errors='coerce' )
FD['S2Prime'] = pd.to_numeric(FD.S2Prime,errors='coerce' )
FD['S3Prime'] = pd.to_numeric(FD.S3Prime,errors='coerce' )
FD['S4Prime'] = pd.to_numeric(FD.S4Prime,errors='coerce' )
```

```python
# For this model we would be replacing the missing values respectively. 
# C1 : Mean ( continuous)
FD['C1'] = FD['C1'].fillna(FD['C1'].mean())
# C4 : Mean (continuos)
FD['C4'] = FD['C4'].fillna(FD['C4'].mean())
# C5prime : Mean (continuous)
#FD['C5Prime'] = FD['C5Prime'].fillna(FD['C5Prime'].mean())
#T1Prime : Mean(continuous)4
FD['T1Prime'] = FD['T1Prime'].fillna(FD['T1Prime'].mean())
#T2Prime : Mean(continuous)
FD['T2Prime'] = FD['T2Prime'].fillna(FD['T2Prime'].mean())
#S1Prime : Mean(continuous)
FD['S1Prime'] = FD['S1Prime'].fillna(FD['S1Prime'].mean())
#S2Prime : Mean(continuous)
FD['S2Prime'] = FD['S2Prime'].fillna(FD['S2Prime'].mean())
#S3Prime : Mean(continuous)
FD['S3Prime'] = FD['S3Prime'].fillna(FD['S3Prime'].mean())
#S4Prime : Mean(continuous)
FD['S4Prime'] = FD['S4Prime'].fillna(FD['S4Prime'].mean())


# C7 : Median (continuous but the amount is in millions hence median would be better)
FD['C7'] = FD['C7'].fillna(FD['C7'].median())


#Filling the categorical data with the majority 
# C2 : Majority (Categorical data : Most frequent)
FD = FD.fillna(FD['C2'].value_counts().index[0])


# Checking for missing values :
FD.isnull().sum(axis=0).tolist()
```

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt 
plt.rc("font", size=14)
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
import seaborn as sns
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
kfold = model_selection.KFold(n_splits=10, random_state=7)
modelCV = LogisticRegression()
scoring = 'accuracy'
results = model_selection.cross_val_score(modelCV, X_train, y_train, cv=kfold, scoring=scoring)
print("10-fold cross validation average accuracy: %.3f" % (results.mean()))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
